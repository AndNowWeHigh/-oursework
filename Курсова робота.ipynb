<h2>4.2</h2><br>
<span style='font-size:18px;'>–í–∏—Ä–∞—Ö—É–≤–∞—Ç–∏ –º–∞—Ç–µ–º–∞—Ç–∏—á–Ω–µ —Å–ø–æ–¥—ñ–≤–∞–Ω–Ω—è —Ç–∞ –¥–∏—Å–ø–µ—Ä—Å—ñ—é –¥–ª—è –Ω–∞–±–æ—Ä—É –¥–∞–Ω–∏—Ö. –°—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑—É–≤–∞—Ç–∏ –¥–∞–Ω—ñ.</span>

$$
\bar{X_{std}} = \frac{1}{\sigma} (X - \mu)
$$

$$
\mu = \sum p_i X_i  {-–¥–µ  p_i-–π–º–æ–≤—ñ—Ä–Ω—ñ—Å—Ç—å \hspace{0.25cm} –ø–æ—è–≤–∏ \hspace{0.25cm} –µ–ª–µ–º–µ–Ω—Ç—É X_i}
$$

$$
\sigma = \sqrt{\frac{\sum_{}{} (X_i - \bar{X})}{n-1}}
$$

```python
# –û–±—á–∏—Å–ª–µ–Ω–Ω—è –º–∞—Ç–µ–º–∞—Ç–∏—á–Ω–æ–≥–æ —Å–ø–æ–¥—ñ–≤–∞–Ω–Ω—è
# data.mean(numeric_only=True)
def calculate_mean(df):
    means = {}
    for col in df.columns:
        if pd.api.types.is_numeric_dtype(df[col]):
            n = len(df[col])
            sum_values = sum(df[col])
            mean_value = sum_values / n
            means[col] = mean_value
    return means
# print(calculate_mean(data))


# –û–±—á–∏—Å–ª–µ–Ω–Ω—è –¥–∏—Å–ø–µ—Ä—Å—ñ—ó
#  variance_values = df.var()
def var(df):
    means = calculate_mean(df)
    var1 = {}
    for col in df.columns:
        if pd.api.types.is_numeric_dtype(df[col]):
            n = len(df[col])
            mean = means[col]
            var1[col] = sum((item - mean)**2 for item in df[col]) / (n - 1)
    return var1
# print(var(data))

#standardized_data = (data - mean_values) / np.sqrt(variance_values)
def standardized_data(df):
    means = calculate_mean(df)
    var1 = var(df)
    copy_df = df.copy()
    for col in df.columns:
        if pd.api.types.is_numeric_dtype(df[col]):
            df[col] = ((df[col] - means[col]) / np.sqrt(var1[col]))
            
    return df

standardized_data = standardized_data(df)    
    
```

–¥–ª—è –æ–±–æ—Ö. –ü—ñ–¥—Ä–∞—Ö—É–≤–∞—Ç–∏ –∫–æ—Ä–µ–ª—è—Ü—ñ—é. –ó–±—É–¥—É–≤–∞—Ç–∏ multipple scatterplot –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é FacetGrid
–¥–ª—è —Ü–∏—Ö –æ–Ω–æ–≤–ª–µ–Ω–∏—Ö –æ–∑–Ω–∞–∫ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—é—á–∏ –±—ñ–±–ª—ñ–æ—Ç–µ–∫—É seaborn –≤–∑—è–≤—à–∏ —Ç—Ä–µ—Ç—é –æ–∑–Ω–∞–∫—É –∑–∞ –≤—ñ—Å—å x.

```python
corr_df = c_data.corr(method="pearson", numeric_only=True)
top_correlations = (corr_df.abs().stack()
                    .sort_values(ascending=False)
                    .reset_index()
                    .rename(columns={0: 'correlation'}))
top_correlations = top_correlations.iloc[12:, :]
# –í–∏–±–µ—Ä–µ–º–æ —Å—Ç–æ–≤–ø—Ü—ñ, —è–∫—ñ –ø—Ä–∏—Å—É—Ç–Ω—ñ –≤ c_data
top_2 = top_correlations.iloc[0, :2].values

# –°—Ç–≤–æ—Ä–∏–º–æ –∫–æ–ø—ñ—é –¥–ª—è –≤—ñ–¥—Ñ—ñ–ª—å—Ç—Ä–æ–≤–∞–Ω–∏—Ö –¥–∞–Ω–∏—Ö
filtered_data = c_data.copy()

# –í–∏–¥–∞–ª–µ–Ω–Ω—è 10% –≤–∏–ø–∞–¥–∫–æ–≤–∏—Ö –∑–Ω–∞—á–µ–Ω—å –¥–ª—è –≤–∫–∞–∑–∞–Ω–∏—Ö –æ–∑–Ω–∞–∫
for feature in top_2:
    mask = np.random.rand(len(filtered_data)) < 0.9
    filtered_data.loc[mask, feature] = np.nan

# –°—Ç–≤–æ—Ä–µ–Ω–Ω—è FacetGrid
g = sns.FacetGrid(filtered_data, col="quality", hue="type")
g.map(plt.scatter, selected_columns[0], selected_columns[1], alpha=0.7)
g.add_legend()

# –í—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω–Ω—è –ø—ñ–¥–ø–∏—Å—ñ–≤ –¥–ª—è –≤–æ—Å—å–º–æ–≥–æ —Å—Ç–æ–≤–ø—Ü—è 
g.set_axis_labels(selected_columns[0], selected_columns[1])

# –í—ñ–¥–æ–±—Ä–∞–∂–µ–Ω–Ω—è –≥—Ä–∞—Ñ—ñ–∫—É
plt.show()
```
